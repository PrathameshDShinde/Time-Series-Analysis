```python
#### order = p,q,d(1,1,1)
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
model = ARIMA(df1['Close'],order = (1,1,1))
model_fit = model.fit()
model_fit.summary()
```

```python
df1['Forecast'] = model_fit.predict(start=200,end=270,dynamic=True)
df1[['Close','Forecast']].plot()
```

```python
import statsmodels.api as sm
model = sm.tsa.statespace.SARIMAX(df1['Close'],order = (1,1,1),seasonal_order=(1,1,1,12))
result = model.fit()
print(result)
```

```python
df1['Forecast'] = result.predict(start=200,end=260,dynamic=True)
df1[['Forecast','Close']].plot()
```
